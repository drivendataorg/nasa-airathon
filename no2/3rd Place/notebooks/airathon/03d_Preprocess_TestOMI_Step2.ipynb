#### User instructions

Run this notebook. It combines all the gridID_testOMI files (e.g., 1X116_testOMI.csv, 1Z2W7_testOMI.csv) into a single testOMI.csv file. This file contains 2 variables: **NO2\_OMI** (concentration in the entire atmospheric column) and **NO2Tr\_OMI** (concentration in the tropospheric column). 

#### Load packages

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from pathlib import Path
import netCDF4
```

#### Directories

```python
ROOT_DIR    = '../../'

#Location of processed datasets
EXTDATA_DIR = ROOT_DIR + 'data/airathon/processed/'
```

#### Load OBS data

```python
df_OBS    = pd.read_csv(EXTDATA_DIR + 'test/STN/' + 'testOBS.csv') #Contains: datetime, ID, latitude, longitude, PM2.5

ID   = df_OBS['ID']
LONx = df_OBS['longitude']
LATx = df_OBS['latitude']

nSamples  = np.size(ID)

YR = (pd.to_datetime(df_OBS['datetime'])).dt.year
MO = (pd.to_datetime(df_OBS['datetime'])).dt.month
DY = (pd.to_datetime(df_OBS['datetime'])).dt.day
```

#### Extract and combine OMI data from each grid ID

```python
cnt = 0
for n in range(nSamples):
    
    #Load OMI csv file for ID[n]
    df_OMI   = pd.read_csv(EXTDATA_DIR + 'test/OMI/' + ID[n] + '_testOMI.csv')
    df       = df_OMI[(df_OMI['Year'] == YR[n]) & (df_OMI['Month'] == MO[n]) & (df_OMI['Day'] == DY[n]) ]

    NO2_OMI      = np.nanmean(df['NO2_OMI'])
    NO2Tr_OMI    = np.nanmean(df['NO2Tr_OMI'])
    
    #Combine variables in an array
    D = np.array((NO2_OMI,NO2Tr_OMI))
    if cnt == 0:
        comboD = D
    else:
        comboD = np.vstack((comboD,D))

    cnt = cnt + 1
            
#Create pandas dataframes and save it as csv files   
df_new = pd.DataFrame(data=comboD)
df_new.columns = ['NO2_OMI','NO2Tr_OMI'] 

df_new.to_csv(EXTDATA_DIR + 'test/OMI/' + 'testOMI.csv', index=False)
```
